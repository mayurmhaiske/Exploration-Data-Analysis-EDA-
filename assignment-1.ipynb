```python
plt.figure(figsize=(12, 8))

for i, col in enumerate(num_cols, 1):
    plt.subplot(2, 4, i)
    sns.histplot(df[col], kde=True, bins=40, color='skyblue')
    plt.title(f'{col.capitalize()} Distribution')
    
    plt.subplot(2, 4, i+4)
    sns.boxplot(x=df[col], color='lightcoral')
    plt.title(f'{col.capitalize()} Boxplot')

plt.tight_layout()
plt.show()
```

- Prints summary stats (mean, median, std, etc.).

- Plots histograms (with KDE) to see data spread and skewness.

```python
plt.figure(figsize=(18,8))
plt.subplot(1, 4, 1)
plt.title('sales')
sns.violinplot(y='sales',data=df,color='red',linewidth=3)
plt.subplot(1, 4, 2)
plt.title('profit')
sns.violinplot(y='profit',data=df,color='green',linewidth=3)
plt.subplot(1, 4, 3)
plt.title('discount')
sns.violinplot(y='discount',data=df,color='blue',linewidth=3)
plt.show()
```

### sales 
- Sales are highly skewed.

- Many low-value sales, few very high-value orders.

- Business relies on volume, not big-ticket orders.

### profit
- Profitability is highly inconsistent.

- There are many negative profit transactions.

- Few extremely profitable orders compensate for losses.

### discount 

- Discounting impacts profit heavily
